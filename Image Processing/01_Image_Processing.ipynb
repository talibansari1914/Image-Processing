## Input Image to Gray & Black&White

```python
import cv2
import matplotlib.pyplot as plt

img_path=input('Enter image Path')

# image reading
img=cv2.imread(img_path)

if img is None:
  print(f'{img_path} or image not found or loaded')
else:
  print(f'Image Loaded from path:{img_path}')

  gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

  # Applying Threshold
  _,br=cv2.threshold(gray,127,255,cv2.THRESH_BINARY)

  # Orginial image
  img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  plt.title('Original Image')
  plt.imshow(img_rgb)
  plt.axis('off')
  plt.show()

  # Gray Scale image
  plt.title('GrayScale Image')
  plt.imshow(gray,cmap='gray')
  plt.axis('off')
  plt.show()

  # Black & White
  plt.title('Black & White Image')
  plt.imshow(br,cmap='gray')
  plt.axis('off')
  plt.show()
```

# Image Menu

```python
import cv2
import matplotlib.pyplot as plt

# User input path for image
img_path=input('Enter image Path:')

# Image Loading
img=cv2.imread(img_path)

# Checking image read or not
if img is None:
  print(f'{img_path} or image not found or loaded')
  exit()

print('Image Loaded Successfull')
print('Shape of the Image:',img.shape)

print('\nSelect Operation on Image-')
print('1.Original Image')
print('2.GrayScale Image')
print('3.Black & white Image')
print('4.Blur Image')
print('5.Edge Detection')
print('6.Rotate Image')
print('7.Resize Image')
print('8.Invert Color of Image')
print('9.Increase Brightness')


choice=int(input('Enter Operation Choice:'))

# ------------------Opertaion----------------------

if choice==1:
  result=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

elif choice==2:
  result=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

elif choice==3:
  gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  _,result=cv2.threshold(gray,127,255,cv2.THRESH_BINARY)

elif choice==4:
  result=cv2.GaussianBlur(cv2.cvtColor(img,cv2.COLOR_BGR2RGB),(15,15),0)

elif choice==5:
  result=cv2.Canny(cv2.cvtColor(img,cv2.COLOR_BGR2RGB),100,200)

elif choice==6:
  result=cv2.rotate(cv2.cvtColor(img,cv2.COLOR_BGR2RGB),cv2.ROTATE_90_CLOCKWISE)

elif choice==7:
  width=int(input('Enter width:'))
  height=int(input('Enter height:'))
  result=cv2.resize(cv2.cvtColor(img,cv2.COLOR_BGR2RGB),(width,height))

elif choice==8:
  result=255-(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

elif choice==9:
  result=cv2.convertScaleAbs(cv2.cvtColor(img,cv2.COLOR_BGR2RGB),alpha=1,beta=50)

else:
  print('Invalid Choice')
  exit()

# Image Show
plt.imshow(result,cmap='gray')
plt.title('Result Image')
plt.axis('off')
plt.show()





```

## Soble Edge Detection

```python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('/content/Rose.jpg')

if img is None:
  print('image is not loaded or not found,Please check it Once')
  exit()

print('Image Loaded Successfully')

# BGR to RGB
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

# changing BGR to Gray
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# sobel Edge Detection for Original Image
osobelx=cv2.Sobel(img_rgb,cv2.CV_64F,1,0)
osobely=cv2.Sobel(img_rgb,cv2.CV_64F,0,1)


# Sobel Edge detection for Gray Image
gsobelx=cv2.Sobel(gray,cv2.CV_64F,1,0)
gsobely=cv2.Sobel(gray,cv2.CV_64F,0,1)

plt.figure(figsize=(15,7))
plt.subplot(1,6,1)

plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,6,2)
plt.imshow(osobelx)
plt.title('OSoble X')
plt.axis('off')

plt.subplot(1,6,3)
plt.imshow(osobely)
plt.title('OSoble Y')
plt.axis('off')

plt.subplot(1,6,4)
plt.imshow(gray,cmap='gray')
plt.title('GraScale Image')
plt.axis('off')

plt.subplot(1,6,5)
plt.imshow(gsobelx,cmap='gray')
plt.title('GSobel X')
plt.axis('off')

plt.subplot(1,6,6)
plt.imshow(gsobely,cmap='gray')
plt.title('GSobel Y')
plt.axis('off')

plt.show()
```

```python
import cv2

cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    edges = cv2.Canny(gray,100,200)

    cv2.imshow("Edges", edges)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
